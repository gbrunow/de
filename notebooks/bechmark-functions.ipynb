## Rosenbrock

**Global Minimum**: $f(\textbf{x}^{\ast})=0$ at $\textbf{x}^{\ast} = (1, â€¦, 1)$

$$f(x_1 \cdots x_n) = \sum_{i=1}^{n-1} [b(x_{i+1} - x_i^2)^2 + (a-x_i)^2]$$ where $a = 1$ and $b = 100$

```python
X = np.arange(-2, 2, 0.1)
Y = np.arange(-2, 2, 0.1)
X, Y = np.meshgrid(X, Y)
Xs = np.array([X, Y])
Z = rosenbrock(Xs)

fig = plt.figure()

ax = fig.gca(projection='3d')
ax.view_init(30, 130)

surf = ax.plot_surface(
    X, Y, Z, cmap=cm.plasma,
    linewidth=0, antialiased=False
)

plt.show()
```

## Rastrigin

**Global Minimum**: $f(\textbf{x}^{\ast})=0$ at $\textbf{x}^{\ast} = (0, 0)$.

$$f(x_1 \cdots x_n)=An + \sum_{i=1}^{n}(x_i^2 - Acos(2\pi x_i))$$ where $A = 10$

```python
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X, Y)
Xs = np.array([X, Y])
Z = rastrigin(Xs)

fig2 = plt.figure()

ax = fig2.gca(projection='3d')
ax.view_init(45, 130)

surf = ax.plot_surface(
    X, Y, Z, cmap=cm.magma,
    linewidth=0, antialiased=False
)

plt.show()
```

## Shubert

**Global Minimum**: $f(\textbf{x}^{\ast})\approx-186.7309$

$$f(\mathbf{x})=f(x_1, ...,x_n)=\prod_{i=1}^{n}{\left(\sum_{j=1}^5{ cos((j+1)x_i+j)}\right)}$$ where $n=2$

```python
Xs = np.mgrid[-2:2:0.1,-2:2:0.1]
x = Xs[0].reshape(-1)
y = Xs[1].reshape(-1)
z = shubert(Xs)

fig3 = plt.figure()

ax = fig3.gca(projection='3d')
ax.view_init(45, 150)

surf = ax.plot_trisurf(
    x, y, z, cmap=cm.magma,
    linewidth=0, antialiased=False
)

plt.show()
```
